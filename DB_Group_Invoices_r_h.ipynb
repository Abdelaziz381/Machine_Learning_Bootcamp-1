```python
# import libraries 

import sqlite3
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# make connection and create engine

engine = sqlite3.connect('Sales.db')
con1 = engine.cursor()
```

```python
g = pd.read_csv("Group_.csv")
g.head()
```

```python
qry1 = """
CREATE TABLE Group_(
     GROUP_ID  integer PRIMARY KEY,
     GROUP_NAME varchar(40)
);

"""

con1.execute(qry1)
engine.commit()
```

```python
v = pd.read_csv("INVOICES_R_H.csv")
v.head()
```

```python
qry = """
CREATE TABLE INVOICES_R_H(
     INVOICES_R_H_ID  integer PRIMARY KEY,
     ACCOUNT_ID integer, 
     EMP_ID1 float, 
     DATE_D date, 
     TIME_T time,
     TOTAL_TOTAL float
);

"""

con1.execute(qry)
engine.commit()
```

### Insert data from csv files into DB

```python
g.to_sql("Group_", engine, if_exists='append', index=False)
```

```python
v.to_sql("INVOICES_R_H", engine, if_exists='append', index=False)
```

## import tables from db into Dataframe
